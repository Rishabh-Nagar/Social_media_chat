## Preparing a Dictionary for every Person

```python
from collections import defaultdict
ppl=defaultdict(list)

for line in content:
    try:
        person = line.split(':')[1][8:]
        print(person)
        text = nltk.sent_tokenize(''.join(line.split(':')[2]))
        print(text)
        ppl[person].extend(text)   # If key exists (person), extend list with value (text),
                                   # if not create a new key, with value added to list
    except:
        print(line)  # in case reading a line fails, examine why
        pass
```

## Downloading the NPS_chat corpus of nltk

```python
import nltk
nltk.download('nps_chat')
```

## Train the Model on Chat Corpus

```python
import nltk
posts = nltk.corpus.nps_chat.xml_posts()

def extract_features(post):
    features = {}
    for word in nltk.word_tokenize(post):
        features['contains({})'.format(word.lower())] = True
    return features

fposts = [(extract_features(p.text), p.get('class')) for p in posts]
test_size = int(len(fposts) * 0.1)
train_set, test_set = fposts[test_size:], fposts[:test_size]
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

## Testing

## Testing on our Dataset

```python
a_x = []
dictx = {}
for i in range(len(x[0])):
    res = classifier.classify(extract_features(x[0][i]))
    if res in a_x:
        dictx[res] = dictx.get(res, 0) + 1
    else:
        a_x.append(res)
        dictx[res] = 1
```

```python
dfx = pd.DataFrame(dictx.items(), columns=['category','count_x'])
dfx.head()
```

### For Rohit

```python
a_y = []
dicty = {}
for i in range(len(y[0])):
    res = classifier.classify(extract_features(y[0][i]))
    if res in a_y:
        dicty[res] = dicty.get(res, 0) + 1
    else:
        a_y.append(res)
        dicty[res] = 1
```

```python
dfy = pd.DataFrame(dicty.items(), columns=['category','count_y'])
dfy.head()
```
